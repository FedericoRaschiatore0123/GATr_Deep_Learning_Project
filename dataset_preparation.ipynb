#Extract files

```python
def extract_files(path, destination):

  if not os.path.exists(destination):
    os.makedirs(destination)

  files = []

  with h5py.File(path, 'r') as file:

      for file_name in file.keys():
          input_file = file[file_name]

          face_data = input_file['face'][:]
          inlet_idcs_data = input_file['inlet_idcs'][:]
          pos_data = input_file['pos'][:]
          pressure_data = input_file['pressure'][:]
          wss_data = input_file['wss'][:]

          with h5py.File( destination + file_name +'.hdf5', 'w') as output_file:
              output_file.create_dataset('face', data=face_data)
              output_file.create_dataset('inlet_idcs', data=inlet_idcs_data)
              output_file.create_dataset('pos', data=pos_data)
              output_file.create_dataset('pressure', data=pressure_data)
              output_file.create_dataset('wss', data=wss_data)

              print(file_name + ".hdf5 created successfully.")

          files.append(file_name + ".hdf5")

  return files
```

```python
single_files = extract_files(single, single_directory)
```

```python
bifurcating_files = extract_files(bifurcating, bifurcating_directory)
```

```python
print("single files number: ", len(single_files))
print("bifurcating files number: ",len(bifurcating_files))
```
